```
import datetime as dt
import calendar
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df['month_int'] = df['Date'].dt.month
df['month'] = df['month_int'].apply(lambda x: calendar.month_abbr[x])
```

```
df_11_2020['Date'] = pd.to_datetime(df_11_2020['Date'], dayfirst=True)
df_11_2020['month_int'] = df_11_2020['Date'].dt.month
df_11_2020['month'] = df_11_2020['month_int'].apply(lambda x: calendar.month_abbr[x])
```

```
df1 = df[['Region','Estimated Unemployment Rate']].groupby('Region').sum().sort_values('Estimated Unemployment Rate', ascending=False)
```

```
df1_11_2020 = df_11_2020[['Region','Estimated Unemployment Rate']].groupby('Region').sum().sort_values('Estimated Unemployment Rate', ascending=False)
```

```
df1.head(10)
```

```
df1_11_2020.head(10)
```

```
fig = plt.figure()
ax0 = fig.add_subplot(1,2,1)
df1[:10].plot(kind='bar',color='green',figsize=(30,5),ax=ax0)
ax0.set_title('Top 10 Reions with highest')
ax0.set_xlabel('Region')
ax0.set_ylabel('Number of people unemployment %')
```

```
fig = plt.figure()
ax0 = fig.add_subplot(1,2,1)
df1_11_2020[:10].plot(kind='bar',color='purple',figsize=(30,5),ax=ax0)
ax0.set_title('Top 10 Reions with highest')
ax0.set_xlabel('Region')
ax0.set_ylabel('Number of people unemployment %')
```

```
df2 = df[['month','Estimated Unemployment Rate']].groupby('month').sum().sort_values(by='Estimated Unemployment Rate', ascending=False)
df2.head(10)
```

```
df2_11_2020 = df[['month','Estimated Unemployment Rate']].groupby('month').sum().sort_values(by='Estimated Unemployment Rate', ascending=False)
df2_11_2020.head(10)
```

```
fig = plt.figure()
ax0 = fig.add_subplot(1,2,1)
df2[:12].plot(kind='bar',color='pink',figsize=(30,5),ax=ax0)
ax0.set_title('Top 10 Reions with highest')
ax0.set_xlabel('Region')
ax0.set_ylabel('Number of people unemployment %')
```

```
fig = plt.figure()
ax0 = fig.add_subplot(1,2,1)
df2_11_2020[:12].plot(kind='bar',color='yellow',figsize=(30,5),ax=ax0)
ax0.set_title('month with highest unemployment')
ax0.set_xlabel('Region')
ax0.set_ylabel('Number of people unemployment (in %)')
```
